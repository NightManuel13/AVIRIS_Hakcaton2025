# [dbo].[fn_CalcularCostoReservacion]

```sql
USE [TurismoAvesDB3]
GO
```

```sql
/****** Object:  UserDefinedFunction [dbo].[fn_CalcularCostoReservacion]    Script Date: 29/9/2025 21:39:07 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE FUNCTION [dbo].[fn_CalcularCostoReservacion]
(
    @IdReserva INT,
    @Fecha DATE,
    @Adultos INT,
    @Ninos INT
)
RETURNS DECIMAL(12,2)
AS
BEGIN
    DECLARE @Costo DECIMAL(12,2);

    SELECT @Costo = (@Adultos * CostoAdulto + @Ninos * CostoNino)
    FROM CupoReserva
    WHERE IdReserva = @IdReserva AND Fecha = @Fecha;

    RETURN ISNULL(@Costo,0);
END;
GO
```

# [dbo].[vw_Reservaciones]

```sql
/****** Object:  View [dbo].[vw_Reservaciones]    Script Date: 29/9/2025 21:39:07 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

-- Vista: Reservaciones con detalle
CREATE   VIEW [dbo].[vw_Reservaciones]
AS
SELECT res.IdReservacion,
       res.ClienteNombre,
       res.ClienteCorreo,
       res.FechaReserva,
       res.CantAdultos,
       res.CantNinos,
       res.CostoTotal,
       res.Estado,
       r.Nombre AS Reserva,
       rt.Nombre AS Ruta,
       g.Nombre + ' ' + g.Apellido AS Guia
FROM Reservacion res
JOIN ReservaNatural r ON res.IdReserva=r.IdReserva
JOIN RutaTuristica rt ON res.IdRuta=rt.IdRuta
LEFT JOIN GuiaTuristico g ON res.IdGuia=g.IdGuia;
GO
```

# [dbo].[vw_AvesPorReserva]

```sql
/****** Object:  View [dbo].[vw_AvesPorReserva]    Script Date: 29/9/2025 21:39:07 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE   VIEW [dbo].[vw_AvesPorReserva]
AS
SELECT 
    rn.Nombre AS Reserva,
    a.NombreComun AS Ave,
    a.NombreCientifico,
    tm.Nombre AS TipoMigracion,
    ISNULL(mr.FechaInicio, ma.FechaInicio) AS FechaInicio,
    ISNULL(mr.FechaFin, ma.FechaFin) AS FechaFin,
    ISNULL(mr.Notas, ma.Notas) AS Notas
FROM MigracionReserva mr
JOIN MigracionAve ma ON mr.IdMigracion = ma.IdMigracion
JOIN Ave a ON ma.IdAve = a.IdAve
JOIN TipoMigracion tm ON ma.IdTipoMigracion = tm.IdTipoMigracion
JOIN ReservaNatural rn ON mr.IdReserva = rn.IdReserva;
GO
```

# [dbo].[sp_CancelarReservacion]

```sql
/****** Object:  StoredProcedure [dbo].[sp_CancelarReservacion]    Script Date: 29/9/2025 21:39:07 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[sp_CancelarReservacion]
    @IdReservacion INT,
    @Motivo NVARCHAR(500) = NULL
AS
BEGIN
    SET NOCOUNT ON;
    BEGIN TRAN;
    BEGIN TRY
        DECLARE @IdReserva INT,@Fecha DATE,@Total INT,@IdCupo INT;

        SELECT @IdReserva=IdReserva,@Fecha=FechaReserva,@Total=TotalPersonas
        FROM Reservacion WHERE IdReservacion=@IdReservacion;

        IF @IdReserva IS NULL
        BEGIN
            ROLLBACK;
            RAISERROR('Reservación no encontrada.',16,1);
            RETURN;
        END

        SELECT @IdCupo=IdCupoReserva
        FROM CupoReserva WITH (UPDLOCK,HOLDLOCK)
        WHERE IdReserva=@IdReserva AND Fecha=@Fecha;

        UPDATE Reservacion
        SET Estado='CANCELADA', Notas=CONCAT(ISNULL(Notas,''),' | CANCELADA: ',@Motivo)
        WHERE IdReservacion=@IdReservacion;

        IF @IdCupo IS NOT NULL
        BEGIN
            UPDATE CupoReserva SET CupoDisponible=CupoDisponible+@Total
            WHERE IdCupoReserva=@IdCupo;

            INSERT INTO LogCupo(IdCupoReserva,TipoCambio,Variacion,Notas)
            VALUES(@IdCupo,'CANCELACION',@Total,CONCAT('Cancelacion=',@IdReservacion));
        END

        COMMIT;
    END TRY
    BEGIN CATCH
        IF XACT_STATE()<>0 ROLLBACK;
        RAISERROR('Error al cancelar la reservación: %s',16,1);
    END CATCH
END;
GO
```

# [dbo].[sp_CrearReservacion]

```sql
/****** Object:  StoredProcedure [dbo].[sp_CrearReservacion]    Script Date: 29/9/2025 21:39:07 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[sp_CrearReservacion]
    @IdReserva INT,
    @IdRuta INT,
    @IdGuia INT = NULL,
    @Fecha DATE,
    @ClienteNombre NVARCHAR(200),
    @ClienteCorreo NVARCHAR(150),
    @Adultos INT,
    @Ninos INT,
    @Notas NVARCHAR(1000) = NULL
AS
BEGIN
    SET NOCOUNT ON;
    DECLARE @TotalPersonas INT = @Adultos + @Ninos;
    DECLARE @IdCupo INT, @Disponibles INT, @Costo DECIMAL(12,2);

    IF @TotalPersonas <= 0
    BEGIN
        RAISERROR('Debe reservar al menos una persona.',16,1);
        RETURN;
    END

    BEGIN TRAN;
    BEGIN TRY
        SELECT @IdCupo = IdCupoReserva, 
               @Disponibles = CupoDisponible
        FROM CupoReserva WITH (UPDLOCK, HOLDLOCK)
        WHERE IdReserva = @IdReserva AND Fecha = @Fecha;

        IF @IdCupo IS NULL
        BEGIN
            ROLLBACK;
            RAISERROR('No existe disponibilidad configurada para esa fecha.',16,1);
            RETURN;
        END

        IF @Disponibles < @TotalPersonas
        BEGIN
            ROLLBACK;
            RAISERROR('No hay cupos suficientes. Disponibles: %d',16,1,@Disponibles);
            RETURN;
        END

        SET @Costo = dbo.fn_CalcularCostoReservacion(@IdReserva,@Fecha,@Adultos,@Ninos);

        INSERT INTO Reservacion (IdReserva, IdRuta, IdGuia, FechaReserva, ClienteNombre, ClienteCorreo,
                                 CantAdultos, CantNinos, CostoTotal, Estado, Notas)
        VALUES (@IdReserva,@IdRuta,@IdGuia,@Fecha,@ClienteNombre,@ClienteCorreo,
                @Adultos,@Ninos,@Costo,'CONFIRMADA',@Notas);

        DECLARE @IdReservacion INT = SCOPE_IDENTITY();

        UPDATE CupoReserva
        SET CupoDisponible = CupoDisponible - @TotalPersonas
        WHERE IdCupoReserva = @IdCupo;

        INSERT INTO LogCupo (IdCupoReserva,TipoCambio,Variacion,Notas)
        VALUES (@IdCupo,'RESERVACION',-@TotalPersonas,CONCAT('Reservacion=',@IdReservacion));

        COMMIT;
        SELECT @IdReservacion AS IdReservacion, @Costo AS CostoTotal;
    END TRY
    BEGIN CATCH
        IF XACT_STATE()<>0 ROLLBACK;
        RAISERROR('Error al crear la reservación: %s',16,1);
    END CATCH
END;
GO
```
